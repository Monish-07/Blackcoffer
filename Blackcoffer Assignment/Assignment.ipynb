```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import nltk
# import syllables
from nltk import sent_tokenize
from nltk import word_tokenize
from nltk.corpus import words
# from syllables import count
import re
```

```python
def extract_data_from_url(url):
    try:
        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            title = soup.title.string if soup.title else 'No title found'

            paragraphs = soup.find_all('p')
            info = ' '.join([p.get_text() for p in paragraphs])

            return title, info, soup

        else:
            return 'Failed to fetch data', 'Failed to fetch data', None
    except Exception as e:
        print(f"An error occurred while fetching data from {url}: {str(e)}")
        return 'Error', 'Error', None

extracted_titles = []
extracted_info = []
extracted_soup = []

for index, row in data.iterrows():
    url = row['URL']
    title, info, soup = extract_data_from_url(url)
    extracted_titles.append(title)
    extracted_info.append(info)
    extracted_soup.append(soup)

data['extracted_titles'] = extracted_titles
data['extracted_info'] = extracted_info
data['extracted_soup'] = extracted_soup

print(data)
```

```python
with open('positive-words.txt', 'r') as file:
    positive_words = set(file.read().splitlines())

with open('negative-words.txt', 'r') as file:
    negative_words = set(file.read().splitlines())

def calculate_sentiment_scores(text):
    words = text.split()
    positive_score = 0 
    negative_score = 0 
    
    for word in words:
        word = word.lower()
        if word not in stop_words:
            if word in positive_words:
                positive_score += 1
            elif word in negative_words:
                negative_score += 1
    return positive_score, negative_score

data['positive_score'], data['negative_score'] = zip(*data['extracted_info'].apply(calculate_sentiment_scores))
data.head()
```

```python
def calculate_polarity_score(row):
    positive_score = row['positive_score']
    negative_score = row['negative_score']
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    return polarity_score

data['polarity_score'] = data.apply(calculate_polarity_score, axis=1)
data.head()
```

```python
data['total_words'] = data['extracted_info'].apply(lambda x: len(x.split()))

def calculate_subjectivity_score(row):
    positive_score = row['positive_score']
    negative_score = row['negative_score']
    total_words = row['total_words']
    
    subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)
    
    return subjectivity_score


data['subjectivity_score'] = data.apply(calculate_subjectivity_score, axis=1)
data.head()
```

```python
english_words = set(words.words())

def calculate_complex_word_count(text):
    words = word_tokenize(text)
    complex_word_count = 0
    for word in words:
        if word.lower() not in english_words:
            complex_word_count += 1
    return complex_word_count

data['complex_word_count'] = data['extracted_info'].apply(calculate_complex_word_count)
data.head(2)
```

```python
def calculate_gunning_fog_index(row):
    average_sentence_length = row['total_words'] / row['sentence_count']
    percentage_of_complex_words = (row['complex_word_count'] / row['total_words']) * 100
    fog_index = 0.4 * (average_sentence_length + percentage_of_complex_words)
    return fog_index

data['gunning_fog_index'] = data.apply(calculate_gunning_fog_index, axis=1)
data.head(2)
```

```python
def calculate_average_words_per_sentence(row):
    return row['total_words'] / row['sentence_count']

data['average_words_per_sentence'] = data.apply(calculate_average_words_per_sentence, axis=1)
data.head(2)
```

```python
def count_syllables(word):
    vowels = "aeiouyAEIOUY"
    count = 0
    prev_char = None
    exceptions = ["es", "ed"]
    
    for char in word:
        if char in vowels:
            if prev_char is None or prev_char not in vowels:
                count += 1
        prev_char = char

    for exception in exceptions:
        if word.endswith(exception):
            count -= 1
    return max(count, 1)

def count_syllables_in_sentence(sentence):
    words = nltk.word_tokenize(sentence)
    syllable_counts = [count_syllables(word) for word in words]
    return sum(syllable_counts)

data['syllable_count'] = data['extracted_info'].apply(count_syllables_in_sentence)
data.head(2)
```

```python
def count_personal_pronouns(text):
    pattern = r'\b(I|we|my|ours|us)\b'
    matches = re.findall(pattern, text, flags=re.IGNORECASE)
    return len(matches)

data['personal_pronoun_count'] = data['extracted_info'].apply(count_personal_pronouns)
data.head(50)
```

```python
def calculate_average_word_length(text):
    words = text.split()
    total_characters = sum(len(word) for word in words) 
    total_words = len(words)
    if total_words > 0:
        average_word_length = total_characters / total_words
    else:
        average_word_length = 0
    return average_word_length

data['average_word_length'] = data['extracted_info'].apply(calculate_average_word_length)

data.head(2)
```

```python
def calculate_average_sentence_length(row):
    return row['total_words'] / row['sentence_count']

data['average_sentence_length'] = data.apply(calculate_average_sentence_length, axis=1)
data.head(2)
```

```python
def calculate_percentage_complex_words(row):
    return (row['complex_word_count'] / row['total_words']) * 100

data['percentage_complex_words'] = data.apply(calculate_percentage_complex_words, axis=1)
data.head(2)
```
